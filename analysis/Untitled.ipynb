```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

train_raw = pd.read_csv('../data/train.csv')
test_set = pd.read_csv('../data/test.csv')

train_features = train_raw.iloc[:,2:]
train_target = LabelEncoder().fit(train_raw.iloc[:,1]).transform(train_raw.iloc[:,1])
train_target = pd.Series(train_target)
test_set = test_set.iloc[:,1:]

print("train_features's shape is {} and its head is \n{}".format(train_features.shape, train_features.head()))
#print("train_target's shape is {} and its head is \n{}".format(train_target.shape, train_target.head()))
print("test_set's shape is {} and its head is \n{}".format(test_set.shape, test_set.head()))
train_target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss


X_train, X_test, y_train, y_test = train_test_split(train_features, train_target, test_size = 0.2, random_state = 123)
from sklearn.svm import SVC
model = SVC(kernel = 'linear', C=1000)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

model.score(X_test, y_test)
```
